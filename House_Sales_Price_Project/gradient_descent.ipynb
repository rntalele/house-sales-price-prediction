```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
from sklearn import linear_model
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
df1=pd.read_csv("progress.csv")
df1.head()
'''df1.drop('X',axis='columns',inplace=True)
df1.head()
df1.drop('Y',axis='columns',inplace=True)
df1.head()'''
```

```python
df2=df1
```

```python
cols=df1.shape[1]
print(cols)
```

```python
X=df1[['Overall_Material']].values
Y=df1['Sale_Price'].values
```

```python
X_train, X_test, Y_train, Y_test = train_test_split(X,Y,test_size=0.2,random_state=0)
```

```python
#create linear regression object
regr= linear_model.LinearRegression()
# Train the model using training sets
regr.fit(X_train, Y_train)
# Make predictions using the testing set
Y_pred=regr.predict(X_test)
```

```python
print(Y_pred)
```

```python
print("Mean squared error: %.2f"% mean_squared_error(Y_test, Y_pred))
print('Regression Score: %.2f'%r2_score(Y_test, Y_pred))
```

```python
sgdReg=linear_model.SGDRegressor()
sgdReg.fit(X_train, Y_train)
Y_pred=sgdReg.predict(X_test)
```

```python
print(Y_pred)
```

```python
%matplotlib inline

fig,ax=plt.subplots(figsize=(12,8))
ax.plot(X_test,Y_pred,label='Prediction',color='red')
ax.scatter(X_train, Y_train, label='Training Data', color='blue')
ax.legend(loc=2)
ax.set_xlabel('Overall Material')
ax.set_ylabel('Sale Price')
ax.set_title('Predicted Overall Material vs Sale Price.')
```
