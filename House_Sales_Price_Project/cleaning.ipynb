```python
X=df[['Overall_Material','Construction_Year','Remodel_Year','Total_Basement_Area','First_Floor_Area','Grade_Living_Area','Garage_Size','Rooms_Above_Grade']].values
y=df['Sale_Price'].values
```

```python
regressor = LinearRegression()  
regressor.fit(X_train, y_train)
```

```python
y_pred = regressor.predict(X_test)
```
